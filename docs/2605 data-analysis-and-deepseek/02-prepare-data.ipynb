```python
movie_id="34780991"
url = f"https://movie.douban.com/subject/34780991/comments?satus=P"

import requests
from bs4 import BeautifulSoup
import pandas as pd

raw_text:str=""

## manual open url in browser

with open('douban-34780991.txt', 'r', encoding='utf-8') as f:
    raw_text = f.read()
```

```python
soup = BeautifulSoup(raw_text, 'html.parser')
# soup

comments_data = []
comments = soup.find_all('div', class_='comment-item')


for comment in comments:
    commenter = comment.find('span', class_='comment-info').find('a').text.strip()
    rating = comment.find('span', class_='rating')['class'][0] if comment.find('span', class_='rating') else 'No rating'
    content = comment.find('span', class_='short').text.strip()

    comments_data.append([commenter,rating, content])
```

```python
df = pd.DataFrame(comments_data, columns=["评论人", "评分", "内容"])
df
```

```python
raw_review:str=""

## manual open url in browser

with open('douban-34780991-review.txt', 'r', encoding='utf-8') as f:
    raw_review = f.read()
```

```python

soup = BeautifulSoup(raw_review, 'html.parser')
# soup

reviews_data = []
reviews = soup.find_all('div', class_='review-item')

for review in reviews:
    # print(review)
    
    reviewer = review.select_one('header a.name').text.strip()
    rating = review.select_one('span.main-title-rating')['title']
    date = review.select_one('span.main-meta').text.strip()

    header = review.select_one('div.main-bd h2')
    title = header.text.strip()
    url = header.select_one('a')['href']
    short_content = review.select_one('div.short-content').text.strip()

    action = review.select_one('div.action')
    up = action.select_one('a.up').text.strip()
    down = action.select_one('a.down').text.strip()

    reviews_data.append([reviewer, rating, date, title,  short_content,  up, down, url])

    
    
```

```python
df = pd.DataFrame(reviews_data, columns=["评论人", "评分", "日期", "标题", "短评", "赞", "踩", "链接"])
df
```
